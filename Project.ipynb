```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("suicide-rate.csv")
print(df.head())
df.head()
```

```python
# Sort data by male suicide rate and select top 10 countries
top_countries = df.sort_values('SuicideRate_Male_RatePer100k_2021', ascending=False).head(10)

plt.figure(figsize=(15, 8))
sns.barplot(data=top_countries, x='country', y='SuicideRate_Male_RatePer100k_2021', color='b', label='Male')
sns.barplot(data=top_countries, x='country', y='SuicideRate_Female_RatePer100k_2021', color='r', label='Female', alpha=0.6)
plt.title("Top 10 Countries by Suicide Rate (per 100,000 people)")
plt.xlabel("Country")
plt.ylabel("Suicide Rate (per 100k)")
plt.legend(title="Gender")

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Female suicidal rate is highest in greenland as compared to others.Mens suicidal rate is lowest Uruguay and Belarus
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select top 5 countries by male suicide rate
top_5_countries = df.nlargest(5, 'SuicideRate_Male_RatePer100k_2021')

# Set 'country' as the index to prepare for the heatmap
df_heatmap = top_5_countries[['country', 'SuicideRate_Male_RatePer100k_2021', 'SuicideRate_Female_RatePer100k_2021']]
df_heatmap = df_heatmap.set_index('country')

# Create the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df_heatmap, annot=True, cmap="YlGnBu", cbar_kws={'label': 'Suicide Rate per 100k'}, linewidths=0.5, annot_kws={"size": 12})

# Add title
plt.title("Top 5 Countries by Suicide Rate (2021)")
plt.xlabel("Gender")
plt.ylabel("Country")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Filter dataset to include only relevant columns
df_pie = df[['country', 'SuicideRate_BothSexes_RatePer100k_2021']]

# Drop any rows with missing values in the relevant columns
df_pie = df_pie.dropna(subset=['SuicideRate_BothSexes_RatePer100k_2021'])

# Convert the suicide rate column to numeric if needed
df_pie['SuicideRate_BothSexes_RatePer100k_2021'] = pd.to_numeric(df_pie['SuicideRate_BothSexes_RatePer100k_2021'], errors='coerce')

# Select the top 5 countries by suicide rate
top_5_countries = df_pie.nlargest(5, 'SuicideRate_BothSexes_RatePer100k_2021')

# Set the country column as labels and suicide rate as values
labels = top_5_countries['country']
sizes = top_5_countries['SuicideRate_BothSexes_RatePer100k_2021']

# Plot pie chart
plt.figure(figsize=(10, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Top 5 Countries by Suicide Rates (Both Sexes, 2021)")
plt.show()
```
